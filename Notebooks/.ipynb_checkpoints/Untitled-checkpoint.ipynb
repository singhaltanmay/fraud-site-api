```python
import pandas as pd
import numpy as np
```

```python
url_df = pd.read_csv('E:\\ML Projects\\ah shit here we go again\\Data\\Datasets\\dataset_phishing.csv')
```

```python
url_df.head()
```

```python
urls=url_df['url'].values
```

```python
unique_chars=[]
maxlen=0
for url in urls:
    maxlen = max(len(url), maxlen)
    for i in url:
        if i in unique_chars:
            continue
        else:
            unique_chars.append(i)
```

```python
unique_chars = sorted(unique_chars)
len(unique_chars), maxlen
```

```python
maxdepth = 10 # 1 se 10 tak
depth=1
#depth=1 means truncate link at length = (depth/maxdepth) * maxlen
length = int((depth/maxdepth) * maxlen)
dic = {}
for i in range(len(unique_chars)):
    dic[unique_chars[i]]=i
    
dic
```

```python
urls_matrix=[]
for url in urls:
    url_vector=[0 for i in range(length)]
    for i in range(length):                                 
        try:
            url_vector[i]=dic[url[i]]
        except IndexError:
            continue
    urls_matrix.append(url_vector)
    
```

```python
urls_matrix = np.array(urls_matrix)
urls_matrix.shape
```

```python
X = urls_matrix
```
